```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
r = requests.get('https://en.wikipedia.org/wiki/Stone_Age')
soup = BeautifulSoup(r.text, 'lxml')
s = soup.find('div', id="bodyContent")
```

```python
'№1 - медленно и неправильный ответ'

i = 2
count = 0
while i < 400:
    if soup.find('img', width=str(i)):
        #print(soup.find('img', width=str(i)))
        count += 1
    i += 1
print(count)
```

```python
'№1'

img = s.find_all('img')
count = 0
for i in range(0, len(img)):
    if int(re.search(r'\swidth=.(\d+).', str(img[i])).group(1)) >= 200:
        count += 1
count
```

```python
'№2'

count = 0
for i in range(1,7):
    text = s.find_all('h' + str(i))
    for txt in text:
        if txt.text[0] in 'ETC': # or txt.text[0] == 'E' or txt.text[0] == 'T' or txt.text[0] == 'C'
            count += 1
count
```
